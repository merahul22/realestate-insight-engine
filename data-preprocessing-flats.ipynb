```python
def treat_price(x):
    if pd.isna(x) or x in ['NaN', 'nan', '', None]:
        return np.nan

    try:
        # Replace non-breaking spaces and trim
        x = str(x).replace('\xa0', ' ').strip().lower()
        parts = x.split(' ')

        if len(parts) < 2:
            return np.nan  # not enough data to parse

        value = float(parts[0].replace(',', ''))
        unit = parts[1]

        if 'crore' in unit:
            return round(value, 2)  # in crores already
        elif 'lac' in unit:
            return round(value / 100, 2)  # convert lac to crore
        elif 'thousand' in unit:
            return round(value / 100000, 4)  # convert thousand to crore
        else:
            return np.nan
    except Exception:
        return np.nan
```

```python
df['price'] = df['price'].apply(treat_price)
```

```python

def clean_price_per_sqft(x):
    try:
        if pd.isna(x):
            return np.nan
        # Extract part before '/' (e.g., ₹ 5,000 from ₹ 5,000/sq.ft.)
        value = str(x).split('/')[0]
        # Remove ₹, commas and spaces
        value = value.replace('₹', '').replace(',', '').strip()
        return float(value)
    except:
        return np.nan
```

```python
df['price_per_sqft'] = df['price_per_sqft'].apply(clean_price_per_sqft)
```

```python
def treat_price(x):
    if isinstance(x, float):  # already float or NaN
        return x
    parts = x.split()
    if len(parts) != 2:
        return np.nan  # invalid format
    value, unit = parts
    try:
        value = float(value)
        if unit == 'Lac':
            return round(value / 100, 2)
        elif unit == 'Crore':
            return round(value, 2)
        else:
            return np.nan  # unknown unit
    except ValueError:
        return np.nan
```
